```python
df = pd.read_csv("shopping_data.csv")
df.head()
```

```python
# Columns
df.columns
```

```python
df.dtypes
```

```python
# Drop null rows
df_shopping = df.dropna()
df_shopping.head()
```

```python
new_df=df_shopping.drop(["CustomerID"],axis=1)
new_df.head()
```

```python
def change(Member):
    if Member == "Yes":
        return 1
    else:
        return 0
    
    
new_df["Card Member"]=new_df["Card Member"].apply(change)
new_df.head()
```

```python
new_df["Annual Income"]=new_df["Annual Income"]/1000
new_df.head()
```

```python
# Saving cleaned data
file_path = "shopping_data_cleaned.csv"
new_df.to_csv(file_path, index=False)
```
